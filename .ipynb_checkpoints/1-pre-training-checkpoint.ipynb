```python
import os
import pickle

def get_files(folder):
    cur_files = []
    for root, dirs, files in os.walk(folder):  
        cur_files = files      
    
    cur_files = [folder+f for f in cur_files if not f[0] == '.']
    return cur_files
```

```python
working_dir = '/Users/ennoh/Desktop/Detector/data/'

raw_training = working_dir+'ALLREPORTS/'
labels_training = working_dir+'ShAReCLEFeHealthAA_1stRoundRS_CLEFPIPEDELIMITED_200_CUIs/'
raw_testing =  working_dir+'ALLREPORTS2/'
#labels_testing =working_dir+'Task2ReferenceStd_CLEFShARe2013Test_StrictAndLenientpipe/'
labels_testing =working_dir+ 'Task2ReferenceStd_CLEFShARe2013Test_TestSpansOnlypipe/'


output = working_dir+'output_2/'
train_bio = output+"train_bio/"
test_bio = output+"test_bio/"
train_tags_dict = output+'train_tags_dict'
test_tags_dict = output+'test_tags_dict'
```

```python
#train_files = [train_bio+f for f in listdir(train_bio) if isfile(join(train_bio, f))]
#test_files = [test_bio+f for f in listdir(test_bio) if isfile(join(test_bio, f))]
train_files = get_files(train_bio)
test_files = get_files(test_bio)
```

```python
#Creating a dictionary for labels
training_label_instances = []
for cur_file in train_files:
    #cur_file = train_bio+cur_file 
    lines = open(cur_file,'r').readlines()
    labels = []
    #print(cur_file)
    for line in lines:
        #print(line)
        parts = line.strip().split('\t')
        labels.append(parts[3])
    training_label_instances.append(labels)
    
count = 0
label_dict = {}
for doc in training_label_instances:
    for word in doc:
        if word not in label_dict:
            label_dict[word] = count
            count+=1
            
len(label_dict)
with open(output+'label_dict.pickle', 'wb') as handle:
    pickle.dump(label_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    

#print(training_label_instances)
print(label_dict)
```

```python
#Combine all files into 1) raw file, 2) index file, and 3) word dict, 4) label dict
def combine_and_write(stage, in_data):
    output_word_file = output+stage+'_docs.pkl'
    output_lable_file = output+stage+'_labels.pkl'
    output_lable_index_file = output+stage+'_label_index.pkl'

    file_instances = []
    label_instances = []
    label_instances_index = []

    for cur_file in in_data:
        print(cur_file)
        lines = open(cur_file,'r').readlines()
        doc = []
        labels = []
        for line in lines:
            parts = line.strip().split('\t')
            doc.append(parts[0])
            labels.append(parts[3])                
        file_instances.append(doc)
        label_instances.append(labels)
        
    print(file_instances[0])
        

    for doc in label_instances:
        new_doc = []
        for word in doc:       
            new_doc.append(label_dict[word])
        label_instances_index.append(new_doc)
        
    #Outputing into files
    fo = open(output_word_file, 'wb')
    pickle.dump(file_instances, fo)
    fo.close()

    fo = open(output_lable_file, 'wb')
    pickle.dump(label_instances, fo)
    fo.close()

    fo = open(output_lable_index_file, 'wb')
    pickle.dump(label_instances_index, fo)
    fo.close()
```

```python
combine_and_write("train",train_files)
combine_and_write("test",test_files)
```
